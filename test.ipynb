```python
from tools.animator import Animator

# result(=PossessionDetector 출력)에서 loss/gain 이벤트 테이블 생성
loss_events = result.loc[result["is_loss"], ["frame_id", "timestamp", "loss_player"]].copy()
loss_events["event_type"] = "loss"
loss_events["player_id"] = loss_events["loss_player"]

gain_events = result.loc[result["is_gain"], ["frame_id", "timestamp", "gain_player"]].copy()
gain_events["event_type"] = "gain"
gain_events["player_id"] = gain_events["gain_player"]

events = pd.concat(
    [
        loss_events[["frame_id", "timestamp", "event_type", "player_id"]],
        gain_events[["frame_id", "timestamp", "event_type", "player_id"]],
    ],
    ignore_index=True,
)

if "period_id" in result.columns:
    period_map = result[["frame_id", "period_id"]].drop_duplicates(subset=["frame_id"])
    events = events.merge(period_map, on="frame_id", how="left")

# animator 내부에서 tracking의 duel 프레임을 읽어 event table에 자동으로 duel을 추가함
animator = Animator(
    track_dict={"main": result.copy()},
    events=events,
    show_events=True,
    play_speed=2,
)
anim = animator.run(max_frames=500, fps=12)
anim.save(os.path.join("possession_events.mp4"), writer="ffmpeg")
anim
```
